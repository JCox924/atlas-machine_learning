# 0-line.py

## Here we create a line plot of a cubic function

```python
import numpy as np
import matplotlib.pyplot as plt

def line():

    x = np.arange(0, 11)
    y = np.arange(0, 11) ** 3
    plt.figure(figsize=(6.4, 4.8))
    plt.plot(x, y, color='#FF0000')
    plt.xlim(0, 10)
    plt.show()

line()
```

# 1-scatter.py

## Here we create a scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

def scatter():

    mean = [69, 0]
    cov = [[15, 8], [8, 15]]
    np.random.seed(5)
    x, y = np.random.multivariate_normal(mean, cov, 2000).T
    y += 180
    plt.figure(figsize=(6.4, 4.8))
    plt.scatter(x, y, color='#FF00FF')
    plt.title("Men's Height vs Weight")
    plt.xlabel('Height (in)')
    plt.ylabel('Weight (lbs)')
    plt.show()

scatter()
```
