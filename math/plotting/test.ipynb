# 0-line.py

## Here we create a line plot of a cubic function

```python
import numpy as np
import matplotlib.pyplot as plt

def line():

    x = np.arange(0, 11)
    y = np.arange(0, 11) ** 3
    plt.figure(figsize=(6.4, 4.8))
    plt.plot(x, y, color='#FF0000')
    plt.xlim(0, 10)
    plt.show()

line()
```

# 1-scatter.py

## Here we create a scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

def scatter():

    mean = [69, 0]
    cov = [[15, 8], [8, 15]]
    np.random.seed(5)
    x, y = np.random.multivariate_normal(mean, cov, 2000).T
    y += 180
    plt.figure(figsize=(6.4, 4.8))
    plt.scatter(x, y, color='#FF00FF', s=10)
    plt.title("Men's Height vs Weight")
    plt.xlabel('Height (in)')
    plt.ylabel('Weight (lbs)')
    plt.show()

scatter()
```

# 2-change_scale.py

```python
import numpy as np
import matplotlib.pyplot as plt

def change_scale():
    x = np.arange(0, 28651, 5730)
    r = np.log(0.5)
    t = 5730
    y = np.exp((r / t) * x)
    plt.figure(figsize=(6.4, 4.8))
    plt.xlim(0, 28651)
    # labels
    plt.title('Exponential Decay of C-14')
    plt.xlabel('Time (years)')
    plt.ylabel('Fraction Remaining')
    # set y axis scale to logarithmic
    plt.yscale('log')
    plt.plot(x, y)



change_scale()
```

# 3-two.py

```python
import numpy as np
import matplotlib.pyplot as plt

def two():

    x = np.arange(0, 21000, 1000)
    r = np.log(0.5)
    t1 = 5730
    t2 = 1600
    y1 = np.exp((r / t1) * x)
    y2 = np.exp((r / t2) * x)
    plt.figure(figsize=(6.4, 4.8))
    plt.title('Exponential Decay of Radioactive Elements')
    plt.xlabel('Time (years)')
    plt.ylabel('Fraction Remaining')
    plt.xlim(0, 20000)
    plt.ylim(0, 1)
    line1, = plt.plot(x, y1, c='#FF0000', ls='--', label='C-14')
    line2, = plt.plot(x, y2, c='g', ls='-', label='Ra-226')
    plt.legend(handles=[line1, line2])
    plt.show()

two()
```

# 4-frequency.py

```python
import numpy as np
import matplotlib.pyplot as plt

def frequency():

    np.random.seed(5)
    student_grades = np.random.normal(68, 15, 50)
    plt.figure(figsize=(6.4, 4.8))

    plt.hist(student_grades, bins=range(0, 101, 10), edgecolor='k')
    plt.title("Project A")
    plt.xlabel('Grades')
    plt.ylabel('Number of Students')
    plt.xlim(0, 100)
    plt.ylim(0, 30)
    max_y = plt.gca().get_ylim()[1]
    plt.yticks(np.arange(0, max_y + 1, 5))
    plt.xticks(np.arange(0, 101, 10))
    
    plt.show()
    
frequency()
```

# 5-all_in_one.py

```python
import numpy as np
import matplotlib.pyplot as plt

def all_in_one():

    y0 = np.arange(0, 11) ** 3

    mean = [69, 0]
    cov = [[15, 8], [8, 15]]
    np.random.seed(5)
    x1, y1 = np.random.multivariate_normal(mean, cov, 2000).T
    y1 += 180

    x2 = np.arange(0, 28651, 5730)
    r2 = np.log(0.5)
    t2 = 5730
    y2 = np.exp((r2 / t2) * x2)

    x3 = np.arange(0, 21000, 1000)
    r3 = np.log(0.5)
    t31 = 5730
    t32 = 1600
    y31 = np.exp((r3 / t31) * x3)
    y32 = np.exp((r3 / t32) * x3)

    np.random.seed(5)
    student_grades = np.random.normal(68, 15, 50)
    
    plt.figure(figsize=(10, 8))
    plt.suptitle('All in One')
    plt.subplot2grid((3, 2), (0, 0))
    x = np.arange(0, 11)
    plt.plot(x, y0, color='#FF0000')
    plt.xlim(0, 10)
    
    plt.subplot2grid((3, 2), (0, 1))
    plt.scatter(x1, y1, color='#FF00FF')
    plt.title("Men's Height vs Weight", fontsize='x-small')
    plt.xlabel('Height (in)', fontsize='x-small')
    plt.ylabel('Weight (lbs)', fontsize='x-small')
    
    plt.subplot2grid((3, 2), (1, 0))
    plt.plot(x2, y2, ls='-', label='C-14')
    plt.xlim(0, 28650)
    plt.yscale('log')
    plt.title('Exponential Decay of C-14', fontsize='x-small')
    plt.xlabel('Time (years)', fontsize='x-small')
    plt.ylabel('Fraction Remaining', fontsize='x-small')
    
    plt.subplot2grid((3, 2), (1, 1))
    plt.title('Exponential Decay of Radioactive Elements', fontsize='x-small')
    plt.xlabel('Time (years)', fontsize='x-small')
    plt.ylabel('Fraction Remaining', fontsize='x-small')
    plt.xlim(0, 20000)
    plt.ylim(0, 1)
    line1, = plt.plot(x3, y31, 'r--', label='C-14')
    line2, = plt.plot(x3, y32, 'g-', label='Ra-226')
    plt.legend(handles=[line1, line2], loc=1)
    
    plt.subplot2grid((3, 2), (2, 0), colspan=2)
    plt.hist(student_grades, bins=range(0, 101, 10), edgecolor='k')
    plt.title("Project A", fontsize='x-small')
    plt.xlabel('Grades', fontsize='x-small')
    plt.ylabel('Number of Students', fontsize='x-small')
    plt.xlim(0, 100)
    plt.ylim(0, 30)
    max_y = plt.gca().get_ylim()[1]
    plt.yticks(np.arange(0, max_y + 1, 5))
    plt.xticks(np.arange(0, 101, 10))
    
    plt.tight_layout()
    plt.show()
    
all_in_one()

    
```

# 6-bars.py

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt

def bars():
    np.random.seed(5)
    fruit = np.random.randint(0, 20, (4,3))
    plt.figure(figsize=(6.4, 4.8))

    names = ['Farrah', 'Fred', 'Felicia']
    
    colors = ['#FF0000', '#FFFF00', '#FF8000', '#FFE5B4']
    
    labels = ['apples', 'bananas', 'oranges', 'peaches']
    
    width = 0.5
    
    indices = np.arange(len(names))
    
    floor = np.zeros(len(names))
    
    for i in range(fruit.shape[0]):
        plt.bar(indices, fruit[i], width=width, bottom=floor, color=colors[i], label=labels[i])
        floor = fruit[i]
        
    plt.ylabel('Quantity of Fruits')
    plt.title('Number of Fruit per Person')
    plt.xticks(indices, names)
    plt.yticks(np.arange(0, 81, 10))
    plt.ylim(0, 80)
    
    plt.legend(loc=1)
    plt.show()
    
bars()
```
